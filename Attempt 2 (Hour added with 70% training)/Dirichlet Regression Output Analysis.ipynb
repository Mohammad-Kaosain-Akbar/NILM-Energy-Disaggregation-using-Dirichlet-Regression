# Here 'Hour' variable was considered while performing the Dirichlet Regression : (Y ~ Total_Value + Hour)

```python
import pandas as pd
import numpy as np
```

# Output after performing Dirichlet Regression

```python
dout = pd.read_csv('DirichletOutput.csv')
dout
```

# Loading test data

```python
dtest = pd.read_csv('DirichletTest.csv')
dtest
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from IPython.display import Image, display

mpl.rcParams['figure.figsize'] = (20,15)
mpl.rcParams['axes.grid'] = False
```

# Analyzing Dirichlet Regression Output with the test value for VENTILATION

```python
vent_out = dout['Ventilation']
vent_out = np.asarray(vent_out).reshape(-1)
vent_test = dtest['Y.Ventilation']
vent_test = np.asarray(vent_test).reshape(-1)
```

```python
plt.plot(vent_out, label='Predicted Value') #plot the loss
plt.plot(vent_test, label='True Value') #plot the validation loss
plt.legend(['Predicted Value', 'True Value'], loc='upper right')
plt.title('Graph for Ventillation Prediction')
```

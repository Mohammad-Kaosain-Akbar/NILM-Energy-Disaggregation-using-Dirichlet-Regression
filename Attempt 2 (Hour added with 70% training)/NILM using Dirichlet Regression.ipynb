```r
library (DirichletReg)
inputData <- read.csv(file = 'dataset.csv')  # plug-in your data here.
set.seed(100)
```

```r
inputData
```

```r
train <- sample (1:nrow (inputData), round (0.7*nrow (inputData)))  # 80% training sample
```

```r
inputData_train <- inputData [train, ] # training Data
# inputData_train
```

```r
inputData_test <- inputData [-train, ] # test Data
# inputData_test
```

```r
inputData$Y <- DR_data (inputData[,6:11])  # prepare the Y's
inputData_train$Y <- DR_data (inputData_train[,6:11])
inputData_test$Y <- DR_data (inputData_test[,6:11])
```

```r
res1 <- DirichReg(Y ~ Total_Value + Hour, data=inputData_train)
```
